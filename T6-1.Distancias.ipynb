# Clustering jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
##Métodos para calcular las distancias linkage, tambien se puede especificar la medida de distancia
##Usando Ward
z = linkage(data[movies],"ward")
z
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el clustering")
plt.xlabel("ID de los usuario de netflix")
plt.ylabel("Distancia")
dendrogram(z, leaf_rotation=90.,leaf_font_size=20.0)
plt.show()
```

```python
##Usando average
z = linkage(data[movies],"average")
z
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el clustering")
plt.xlabel("ID de los usuario de netflix")
plt.ylabel("Distancia")
dendrogram(z, leaf_rotation=90.,leaf_font_size=20.0)
plt.show()
```

```python
##Usando complete
z = linkage(data[movies],"complete")
z
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el clustering")
plt.xlabel("ID de los usuario de netflix")
plt.ylabel("Distancia")
dendrogram(z, leaf_rotation=90.,leaf_font_size=20.0)
plt.show()
```

```python
##Usando single
z = linkage(data[movies],"single")
z
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el clustering")
plt.xlabel("ID de los usuario de netflix")
plt.ylabel("Distancia")
dendrogram(z, leaf_rotation=90.,leaf_font_size=20.0)
plt.show()
```
